```python
import cv2
import numpy as np
from collections import deque
#eline mavi bir nesne alarak yazılan kodu çalıştırabilirsin

cap = cv2.VideoCapture(0)#goruntu alma

#çizim yapacağımız nesneyi aramak için lower ve upper oluşturuyoruz
lower_blue= np.array([100,60,60])
upper_blue= np.array([140,255,255])

#renkleri saklayacağım listeyi oluşturuyorum
blue_points = [deque(maxlen=512)]
green_points = [deque(maxlen=512)]
red_points = [deque(maxlen=512)]
yellow_points = [deque(maxlen=512)]

#renklerin indexleri
blue_index=0
green_index=0
red_index=0
yellow_index=0

#colors içerisinde tuttuğum renk değerleri
colors = [(255,0,0),(0,255,0),(0,0,255),(0,255,255)]
color_index = 0 #varsayılan index yani ilk açıldığında mavi yazar

#paint ekran özelliği
paintWindow = np.zeros((471,636,3))+255

#yapılan çizimleri paintWindow içerisinde saklayacağım bazı optimizasyon sorunlarının önüne geçebilmek için
paintWindow = cv2.rectangle(paintWindow,(40,1),(140,65),(0,0,0),2)
paintWindow = cv2.rectangle(paintWindow,(160,1),(255,65),colors[0],-1)
paintWindow = cv2.rectangle(paintWindow,(275,1),(370,65),colors[1],-1)
paintWindow = cv2.rectangle(paintWindow,(390,1),(485,65),colors[2],-1)
paintWindow = cv2.rectangle(paintWindow,(505,1),(600,65),colors[3],-1)


#font ve ekranda bulunan kutucukların üstüne yazılan yazılar
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(paintWindow,"CLEAR ALL",(49,33),font,0.5,(0,0,0),2,cv2.LINE_AA)
cv2.putText(paintWindow,"BLUE",(185,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paintWindow,"GREEN",(298,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paintWindow,"RED",(420,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paintWindow,"YELLOW",(520,33),font,0.5,(255,255,255),2,cv2.LINE_AA)

cv2.namedWindow("Paint")


while 1:
    ret,frame = cap.read()#her bir frame okuyorum
    frame = cv2.flip(frame,1)#ters cevir
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)#bgr to hsv format

    #renklerin konumlarını belirten kutucuklar
    frame = cv2.rectangle(frame,(40,1),(140,65),(0,0,0),2)
    frame = cv2.rectangle(frame,(160,1),(255,65),colors[0],-1)
    frame = cv2.rectangle(frame,(275,1),(370,65),colors[1],-1)
    frame = cv2.rectangle(frame,(390,1),(485,65),colors[2],-1)
    frame = cv2.rectangle(frame,(505,1),(600,65),colors[3],-1)

    
    #font ve ekranda bulunan kutucukların üstüne yazılan yazılar
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,"CLEAR ALL",(49,33),font,0.5,(0,0,0),2,cv2.LINE_AA)
    cv2.putText(frame,"BLUE",(185,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"GREEN",(298,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"RED",(420,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"YELLOW",(520,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    
    #frameler yanlış okunduğunda kapat
    if ret is False:
        break

    #nesneye maske uyguluyoruz
    mask = cv2.inRange(hsv,lower_blue,upper_blue)
    
    #temiz görüntü oluşturabilmek için mask işlemler uyguluyoruz
    mask = cv2.erode(mask,(5,5),iterations =2)
    mask = cv2.morphologyEx(mask,cv2.MORPH_OPEN,(5,5))
    mask = cv2.dilate(mask,(5,5),iterations = 1)#kalınlaştırıyoruz

    #konturları buluyoruz
    contours,_ =cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    center = None

    #if bloğu oluşturarak renkleri alıyorum
    if len(contours) > 0:
        max_contours = sorted(contours, key = cv2.contourArea, reverse=True)[0]#goruntudeki en buyuk konturları bulma
        ((x,y),radius) = cv2.minEnclosingCircle(max_contours)
        cv2.circle(frame,(int(x),int(y)),int(radius),(255,255,0),3)#sınırlayıcı bir çember çiziyoruz

        M = cv2.moments(max_contours)#merkez bulma
        center = (int(M["m10"]/M["m00"]),int(M["m01"]/M["m00"]))#x ve y değerlerini hesaplıyoruz

        if center[1] <= 65:
            #clear düğmesi var ve tüm değerler 0a eşitlenir
            if 40<=center[0]<=140:

                blue_points = [deque(maxlen=512)]
                green_points = [deque(maxlen=512)]
                red_points = [deque(maxlen=512)]
                yellow_points = [deque(maxlen=512)]

                blue_index=0
                green_index=0
                red_index=0
                yellow_index=0
                
                #tüm ekran tekrardan beyaz olacak
                paintWindow[67:,:,:]=255
                
            #mavi kalem
            elif 160<=center[0]<=255:
                color_index = 0
                
            #yeşil kalem
            elif 275<=center[0]<=370:
                color_index = 1
                
            #kırmızı kalem
            elif 390<=center[0]<=485:
                color_index = 2
            
            #sarı kalem
            elif 505<=center[0]<=600:
                color_index = 3

        #eğer y değeri 65 ten buyukse çizim yapmaya başlıyoruz
        else:
            
            #mavi
            if color_index == 0:
                blue_points[blue_index].appendleft(center)#noktaları ekle(çiz)
            
            #yeşil 
            elif color_index == 1:    
                green_points[green_index].appendleft(center)
                
            #kırmızı 
            elif color_index == 2:
                red_points[red_index].appendleft(center)
             
            #sarı 
            elif color_index == 3:
                yellow_points[yellow_index].appendleft(center)
                
    #indeksleri birer birer arttırıcaz
    else:
        blue_points.append(deque(maxlen=512))
        blue_index+=1
        
        green_points.append(deque(maxlen=512))
        green_index+=1
        
        red_points.append(deque(maxlen=512))
        red_index+=1
        
        yellow_points.append(deque(maxlen=512))
        yellow_index+=1

    points = [blue_points,green_points,red_points,yellow_points]#tüm çizilen değerleri tuttuğumuz
    
    #tüm çizilen değerleri tutalım
    for i in range(len(points)):
        for j in range(len(points[i])):
            for k in range(1,len(points[i][j])):
                if points[i][j][k-1] is None or points[i][j][k] is None:
                    continue
                
                cv2.line(frame,points[i][j][k-1], points[i][j][k], colors[i], 2)
                cv2.line(paintWindow,points[i][j][k-1], points[i][j][k], colors[i], 2)
                
                
    # işlediğimiz kareleri görelim
    cv2.imshow("Frame",frame)
    cv2.imshow("Paint",paintWindow)
    
   # Programı kapatacak kodu yazalım.
    if cv2.waitKey(3) & 0xFF == ord('q'):
        break
        
#videoyu serbest bırakalım
cap.release()
cv2.destroyAllWindows()
```
