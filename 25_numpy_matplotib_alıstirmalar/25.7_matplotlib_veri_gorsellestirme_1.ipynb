```python
#elimizdeki verileri daha anlamlı hale(grafiklere) dönüştürmek için matplotlib kütüphanesi kullanılır
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5)#arange fonksiyonu range gibidir 0-5e kadar sayı üretir[0,1,2,3,4]
y = x

plt.plot(x,y,"o--")  # o, o-, o--
plt.plot(x,-y)
plt.plot(-x,y,"o")
plt.title("y=x, y=-x")#grafiğin adı
plt.show()#goster
```
