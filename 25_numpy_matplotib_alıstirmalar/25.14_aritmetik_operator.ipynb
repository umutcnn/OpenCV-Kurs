```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread("map.jpeg")#resim ekle

plt.subplot(4,2,1)
plt.title("Original Image")#resim balk
plt.imshow(img)#goster

plt.subplot(4,2,2)
plt.title("img+img")#resim balk
plt.imshow(img+img)#goster

plt.subplot(4,2,3)
plt.title("img-img")#resim balk
plt.imshow(img-img)#goster

plt.subplot(4,2,4)
plt.title("np.flip(img,0)")#resim balk
plt.imshow(np.flip(img,0)) #0,1,2

plt.subplot(4,2,5)
plt.title("np.flip(img,1)")#resim balk
plt.imshow(np.flip(img,1)) #0,1,2


plt.subplot(4,2,6)
plt.title("np.flip(img,2)")#resim balk
plt.imshow(np.flip(img,2)) #0,1,2

plt.subplot(4,2,7)
plt.title("np.fliplr(img)") # left to right
plt.imshow(np.fliplr(img)) 

plt.subplot(4,2,8)
plt.title("np.flipud(img)") # updown
plt.imshow(np.flipud(img)) #goster

plt.show()
```
